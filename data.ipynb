```python
def combine_csv(file_list, output_file):
    """
    Combines multiple CSV files into a single CSV file.

    Args:
        file_list (list): A list of paths to the CSV files to be combined.
        output_file (str): The path to the output CSV file.
    """
    combined_df = pd.DataFrame()
    for file in file_list:
        df = pd.read_csv(file)
        combined_df = pd.concat([combined_df, df], ignore_index=True)#just like sum+=num[i]
    combined_df.to_csv(output_file, index=False)


files_to_combine = ['A1/2023_06_10/101_Study_Desk.csv', 'A1/2023_06_11/101_Study_Desk.csv','A1/2023_06_12/101_Study_Desk.csv',
                    'A1/2023_06_13/101_Study_Desk.csv',
                    'A1/2023_06_14/101_Study_Desk.csv','A1/2023_06_16/101_Study_Desk.csv']
combine_csv(files_to_combine, 'A1_Combined.csv')
```

```python
def resample_df(r1):
    r1['ts'] = pd.to_datetime(r1['ts'])
    r1= r1.set_index('ts').sort_index()
    

    # Group into 5-minute intervals
    windowed_df = r1.resample('5Min').agg({
        'Loc':['first'],
        'T':['mean'],
        'H':['mean'],
        'PMS2_5': ['mean'],
        'PMS10':['mean'],
        'CO2':['mean'],
        'NO2':['mean'],
        'VoC': ['mean'],
        'CO': ['mean']
    
    })

    # Flatten multi-index columns
    windowed_df.columns = ['_'.join(col) for col in windowed_df.columns]
    # print(['_'.join(col) for col in windowed_df.columns])
    windowed_df = windowed_df.dropna()
    windowed_df=windowed_df.reset_index()

    return windowed_df  #
```

# Data prep for LSTM

```python
#for each 30 min time frame predicting the next 5 min time frame
#6 values to predict 1 value
def create_sequences(input_data, n_steps, fut_5_mins,):
    X, y = [], []
    for i in range(len(input_data) - n_steps - fut_5_mins):
        # Define the end of the input sequence
        end_ix = i + n_steps
        # Define the end of the output sequence
        out_end_ix = end_ix + fut_5_mins
        # Ensure we have enough data for this instance
        if out_end_ix > len(input_data):
            break
                                 #taking 6 rows,13 cols       #only one row,6 cols
        seq_x, seq_y = input_data[i:end_ix, :], input_data[out_end_ix - 1, 2:8]  # pm25 to No2
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
n_steps=6
fut_5_mins=1
```

```python
X_train, y_train = create_sequences(data_train_normalized, 
                                    n_steps, 
                                    fut_5_mins)
X_val, y_val = create_sequences(data_val_normalized, 
                                n_steps, 
                                fut_5_mins)
X_test, y_test = create_sequences(data_test_normalized, 
                                  n_steps, 
                                  fut_5_mins)
```

# Training

```python
model = Sequential([
    LSTM(128,
         input_shape=(n_steps, X_train.shape[2])),
    Dense(y_train.shape[1])
])
```

```python
model.compile(optimizer='adam', loss='mae')
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    mode='min',
    restore_best_weights=True)
```

restore_best_weights: When set to True, this option restores model weights from the epoch with the best value of the monitored metric ('val_loss' in this case). This means that we will keep the model's best state even if the model's performance degrades in the epochs following the best one (within the patience period).

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=20,
    batch_size=32,
    callbacks=[early_stopping]
)
```

```python
#Converting to tflite without quantization

run_model = tf.function(lambda x: model(x))
BATCH_SIZE = 1
STEPS = n_steps
INPUT_SIZE = 13
#preparing the model for quantization
concrete_func = run_model.get_concrete_function(
    tf.TensorSpec([BATCH_SIZE, STEPS, INPUT_SIZE], model.inputs[0].dtype))

# model directory.
MODEL_DIR = "keras_lstm"
model.save(MODEL_DIR, save_format="tf", signatures=concrete_func)#

converter = tf.lite.TFLiteConverter.from_saved_model(MODEL_DIR)
tflite_model = converter.convert()

# Save the converted model to file
tflite_model_file = 'converted_model.tflite'
with open(tflite_model_file, 'wb') as f:
    f.write(tflite_model)
```

 run_model = tf.function(lambda x: model(x))
This wraps your model in a tf.function.

tf.function converts Python code into a TensorFlow graph (GraphDef) for

Faster execution

Easier export (e.g., TensorFlow Serving, TFLite).

The lambda x: model(x) means:

When you call run_model(input), it passes input directly into model.
